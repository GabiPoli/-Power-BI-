On this notebook I generated the random numbers to use on the Random # spreadsheet and from it I to created the Dashboard in Power BI 

```python
import pandas as pd
import random

#creating the random prices
random_float_list = []
# Set a length of the list to 10
for i in range(0, 1000):
    # any random float between 1.50 to 19.50
    # don't use round() if you need number as it is
    x = round(random.uniform(1.50, 19.50), 2)
    random_float_list.append(x)

print(random_float_list)
```

```python
#saving the prices in a excel format
df = pd.DataFrame()
#creating a column
df['Price'] = random_float_list[0::2]
# Converting to excel
df.to_excel('price.xlsx', index = False)
```

```python
# creating the random strategies
list = ['Pullback', 'Ascending Triangle', 'Dip Buy', 'Cup and Handle']
strategy=random.choices(list, k=1000)
print(random.choices(list, k=1000))
```

```python
#saving the strategy in a excel format
df = pd.DataFrame()
#creating a column
df['Strategy'] = strategy[0::2]
# Converting to excel
df.to_excel('strategy.xlsx', index = False)
```

```python
#creating the difference of prices
random_diff_list = []
# Set a length of the list to 10
for i in range(0, 1000):
    # any random float between -0.5 to 1
    # don't use round() if you need number as it is
    x = round(random.uniform(-0.50, 1), 2)
    random_diff_list.append(x)

print(random_diff_list)
```

```python
#saving the diffprice in a excel format
df = pd.DataFrame()
#creating a column
df['pricediff'] = random_diff_list[0::2]
# Converting to excel
df.to_excel('price diff.xlsx', index = False)
```

```python
#creating a random position
random_position_list = []
# Set a length of the list to 10
for i in range(0, 1000):
    # any random float between 50 to 300
    # don't use round() if you need number as it is
    x = round(random.uniform(50, 300),0)
    random_position_list.append(x)

print(random_position_list)
```

```python
#saving the position in a excel format
df = pd.DataFrame()
#creating a column
df['position'] = random_position_list[0::2]
# Converting to excel
df.to_excel('position.xlsx', index = False)
```
